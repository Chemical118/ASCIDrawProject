```python
import h5py
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
hf = h5py.File("Save\\f4.jld2", "r")
```

```python
m = hf.get('m')[()]
m1 = hf.get('m1')[()]
```

```python
m = np.transpose(m / [np.min(v) for v in np.transpose(m)])
m1 = np.transpose(m1 / [np.min(v) for v in np.transpose(m1)])
```

```python
rel = [65, 66, 203, 205, 222]
data = {
    'val': np.hstack(m),
    'type': ['a', 'b', 'c'] * len(m),
    'residue' : [i for i in rel for _ in range(3)],
}
data1 = {
    'val': np.hstack(m1),
    'type': ['a', 'b', 'c'] * len(m1),
    'residue' : [i for i in rel for _ in range(3)],
}

df = pd.DataFrame(data)
df1 = pd.DataFrame(data1)
```

```python
xlabel = 'avGFP residue'
ylabel = 'Relative importance'
plt.rcParams["font.family"] = "Arial"
```

```python
b = sns.barplot(data=df, x='residue', y='val', hue='type')
b.set_xlabel(xlabel)
b.set_ylabel(ylabel)
b.set_title('avGFPs_em')
plt.savefig("Fig/f4-1.tiff", dpi=600, bbox_inches="tight")
```

```python
b = sns.barplot(data=df1, x='residue', y='val', hue='type')
b.set_xlabel(xlabel)
b.set_ylabel(ylabel)
b.set_title('avGFPs_ex')
plt.savefig("Fig/f4-2.tiff", dpi=600, bbox_inches="tight")
```
